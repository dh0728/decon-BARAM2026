## 6. 기상 예보 데이터 공통 설명

대상 파일:

- `ldaps_train.csv`
- `ldaps_test.csv`

기상 예보 데이터는 매일 `09:00 KST`에 초기화된 예보자료를 사용합니다.

이 예보자료 중 다음날 `01:00:00`부터 그 다음날 `00:00:00`까지의 24시간 예보값만 추출해 날짜별로 이어 붙였습니다.

각 예보자료는 해당일 `13:00 KST`부터 사용 가능한 것으로 간주합니다.

| 컬럼 | 의미 |
|---|---|
| `forecast_kst_dtm` | 예보 대상 시각입니다. 발전량을 예측해야 하는 시각과 대응됩니다. |
| `data_available_kst_dtm` | 해당 예보 데이터가 사용 가능해진 시각입니다. |
| `grid_id` | 기상 격자 ID입니다. |
| `latitude` | 격자 위도입니다. |
| `longitude` | 격자 경도입니다. |

예:

| `forecast_kst_dtm` | `data_available_kst_dtm` |
|---|---|
| `2025-01-01 01:00:00` | `2024-12-31 13:00:00` |
| `2025-01-01 02:00:00` | `2024-12-31 13:00:00` |
| `2025-01-02 00:00:00` | `2024-12-31 13:00:00` |
| `2025-01-02 01:00:00` | `2025-01-01 13:00:00` |

즉 `01:00:00`부터 다음날 `00:00:00`까지의 24개 예보 대상 시각은 같은 `data_available_kst_dtm`을 갖습니다.

평가 기간의 실제 발전량, SCADA, 비공개 운영 데이터, 사후 보정자료, 재분석자료 등은 예측에 사용할 수 없습니다.

Excel에서 CSV 파일을 열고 다시 저장하면 시간 문자열 형식이 바뀔 수 있습니다. 제출 파일은 코드로 생성하는 것을 권장합니다.


```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
DATA_DIR = Path("../datasets")
TRAIN_DIR = DATA_DIR / "train"
TEST_DIR = DATA_DIR / "test"

TARGET_COLS = ["kpx_group_1", "kpx_group_2", "kpx_group_3"]
CAPACITY_KWH = {
    "kpx_group_1": 21600,
    "kpx_group_2": 21600,
    "kpx_group_3": 21000,
}

train_labels = pd.read_csv(TRAIN_DIR / "train_labels.csv", encoding="utf-8-sig")
sample_submission = pd.read_csv(DATA_DIR / "sample_submission.csv", encoding="utf-8-sig")

ldaps_train = pd.read_csv(TRAIN_DIR / "ldaps_train.csv", encoding="utf-8-sig")
gfs_train = pd.read_csv(TRAIN_DIR / "gfs_train.csv", encoding="utf-8-sig")
ldaps_test = pd.read_csv(TEST_DIR / "ldaps_test.csv", encoding="utf-8-sig")
gfs_test = pd.read_csv(TEST_DIR / "gfs_test.csv", encoding="utf-8-sig")

scada_vestas_train = pd.read_csv(TRAIN_DIR / "scada_vestas_train.csv", encoding="utf-8-sig")
scada_unison_train = pd.read_csv(TRAIN_DIR / "scada_unison_train.csv", encoding="utf-8-sig")

# info.xlsx: 상단 3행은 제목/공백이라 스킵하고, 병합 셀로 비어있는 KPX그룹/그룹설비용량(MW)은 ffill로 채움
info_df = pd.read_excel(DATA_DIR / "info.xlsx", skiprows=3)
info_df = info_df.dropna(axis=1, how="all").ffill()

train_labels["kst_dtm"] = pd.to_datetime(train_labels["kst_dtm"])
sample_submission["forecast_kst_dtm"] = pd.to_datetime(sample_submission["forecast_kst_dtm"])

for _df in (ldaps_train, gfs_train, ldaps_test, gfs_test):
    _df["forecast_kst_dtm"] = pd.to_datetime(_df["forecast_kst_dtm"])
    _df["data_available_kst_dtm"] = pd.to_datetime(_df["data_available_kst_dtm"])

for _df in (scada_vestas_train, scada_unison_train):
    _df["kst_dtm"] = pd.to_datetime(_df["kst_dtm"])

print("train_labels:", train_labels.shape)
print("sample_submission:", sample_submission.shape)
print("ldaps_train:", ldaps_train.shape, "gfs_train:", gfs_train.shape)
print("ldaps_test:", ldaps_test.shape, "gfs_test:", gfs_test.shape)
print("scada_vestas_train:", scada_vestas_train.shape)
print("scada_unison_train:", scada_unison_train.shape)
print("info_df:", info_df.shape)
```

## 공통 이슈파악 함수 정의

이후 각 데이터셋(LDAPS/GFS 예보, SCADA)에 반복 적용할 EDA 유틸리티 함수입니다.

```python
def summarize_basic(df: pd.DataFrame, name: str) -> None:
    print(f"===== {name} 기본 EDA =====")
    print("데이터 크기:", df.shape)
    display(df.head())
    print("[dtypes]")
    display(df.dtypes)

    print("[결측치 개수 / 비율(%) - 결측 존재 컬럼만]")
    na_summary = pd.concat(
        [df.isnull().sum(), (df.isnull().mean() * 100).round(2)],
        axis=1, keys=["count", "pct"],
    )
    display(na_summary[na_summary["count"] > 0])
    print("[중복 행 개수]", df.duplicated().sum())

    numeric_cols = df.select_dtypes(include="number").columns
    if len(numeric_cols) == 0:
        return

    print("[숫자형 컬럼 스케일 확인]")
    display(df[numeric_cols].describe().T)

    Q1 = df[numeric_cols].quantile(0.25)
    Q3 = df[numeric_cols].quantile(0.75)
    IQR = Q3 - Q1
    outlier_count = (
        (df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))
    ).sum()
    print("[IQR 기준 이상치 후보 개수 (상위 10개 컬럼)]")
    display(outlier_count.sort_values(ascending=False).head(10))


def check_forecast_grid(df: pd.DataFrame, name: str, expected_grid_count: int) -> None:
    print(f"===== {name} 격자/시각 정합성 =====")
    grid_ids = sorted(df["grid_id"].unique())
    print(f"grid_id 개수: {len(grid_ids)} (기대값 {expected_grid_count})")

    per_time_grid_count = df.groupby("forecast_kst_dtm")["grid_id"].nunique()
    bad_times = per_time_grid_count[per_time_grid_count != expected_grid_count]
    print(f"grid_id 개수가 기대값과 다른 forecast_kst_dtm 시각 수: {len(bad_times)}")
    if len(bad_times):
        display(bad_times.head())

    dup_key = df.duplicated(subset=["forecast_kst_dtm", "grid_id"]).sum()
    print("(forecast_kst_dtm, grid_id) 조합 중복 개수:", dup_key)

    latlon_nunique = df.groupby("grid_id")[["latitude", "longitude"]].nunique()
    moving_grids = latlon_nunique[(latlon_nunique["latitude"] > 1) | (latlon_nunique["longitude"] > 1)]
    print("격자별 좌표가 고정되지 않은 grid_id 개수:", len(moving_grids))

    full_hours = pd.date_range(df["forecast_kst_dtm"].min(), df["forecast_kst_dtm"].max(), freq="h")
    missing_hours = full_hours.difference(df["forecast_kst_dtm"].unique())
    print(f"기간 내 누락된 forecast_kst_dtm 시각 수: {len(missing_hours)} / 전체 {len(full_hours)}")

    block_sizes = df.groupby(["grid_id", "data_available_kst_dtm"])["forecast_kst_dtm"].count()
    print("[data_available_kst_dtm 블록 크기 분포 (정상이면 24시간 단위)]")
    display(block_sizes.value_counts().sort_index())


def check_scada(df: pd.DataFrame, name: str, prefix: str, turbine_ids: list, freq_minutes: int = 10) -> None:
    print(f"===== {name} SCADA 이슈파악 =====")
    dt_col = "kst_dtm"
    print("타임스탬프 중복 개수:", df[dt_col].duplicated().sum())

    full_range = pd.date_range(df[dt_col].min(), df[dt_col].max(), freq=f"{freq_minutes}min")
    missing_ts = full_range.difference(df[dt_col])
    print(f"기간 내 누락된 타임스탬프 수: {len(missing_ts)} / 전체 {len(full_range)}")

    rows = []
    for tid in turbine_ids:
        power_col = f"{prefix}_{tid}_power_kw10m"
        ws_col = f"{prefix}_{tid}_ws"
        wd_col = f"{prefix}_{tid}_wd"
        power = df[power_col]

        same_as_prev = power.notna() & (power == power.shift())
        max_run = same_as_prev.groupby((~same_as_prev).cumsum()).transform("size").where(same_as_prev).max()

        rows.append({
            "turbine": tid,
            "결측치": power.isnull().sum(),
            "음수 power": int((power < 0).sum()),
            "동일값 최대 연속 길이": 0 if pd.isna(max_run) else int(max_run) + 1,
            "풍속0인데 power>0 건수": int(((df[ws_col] == 0) & (power > 0)).sum()),
            "풍향 범위": f"{df[wd_col].min():.1f} ~ {df[wd_col].max():.1f}",
        })
    display(pd.DataFrame(rows).set_index("turbine"))
```

## 이슈파악 요약

- **LDAPS/GFS 예보**: 결측치 없음, grid_id 개수(16/9) 및 좌표 고정 정상, `forecast_kst_dtm` 누락 없음, `data_available_kst_dtm` 24시간 블록 규칙 정상, train/test 컬럼 스키마 일치.
- **train_labels**: `kpx_group_3`이 그룹 설비용량(21,000 kWh)을 초과하는 값이 38건 존재 → 후처리(스케일 보정)나 원인 분석이 필요한 이상치 후보.
- **scada_vestas_train**: `power_kw10m` 컬럼에 정격출력(3.6 MW)로는 설명되지 않는 초대형 이상치(최대 약 ±5×10⁷)가 다수(1~2만 건/터빈) 존재. 정상 범위를 크게 벗어난 값이라 센서 오류/에러 코드일 가능성이 높고, 그대로 사용하면 학습에 치명적 → 클렌징(예: 정격출력 기준 클리핑 또는 결측 처리) 필수.
- **scada_unison_train**: 컬럼별 결측치 0.1~1.4% 존재, `wtg02_power`에 이상치 다수, 풍향(`wd`) 범위가 VESTAS(0~359)와 달리 -180~180로 스케일이 달라 통합 시 변환 필요.
- **info.xlsx**: KPX그룹별 터빈 수(6/6/5)·그룹설비용량(21.6/21.6/21.0 MW)이 `data_description.md` 및 `TARGET_COLS` 설비용량과 일치함을 확인.
